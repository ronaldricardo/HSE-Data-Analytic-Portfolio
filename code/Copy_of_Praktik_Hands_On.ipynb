## Line Chart
Tren insiden bulanan/tahunan Total Recordable Incident Rate (TRIR) & Severity Rate (SR)

```python
# Persiapkan Data TRIR & Hitung TRIR

# 1. Calculate Total_Insiden (Total Recordable Incidents) from df_insiden
df_incidents_monthly = df_insiden.groupby(['Tahun', 'Bulan']).size().reset_index(name='Total_Insiden')

# 2. Merge with Total_Jam_Kerja from df_kpi
df_trir_calculated = pd.merge(
    df_incidents_monthly,
    df_kpi[['Tahun', 'Bulan', 'Total_Jam_Kerja']],
    on=['Tahun', 'Bulan'],
    how='left'
)

# 3. Calculate TRIR: (Total Recordable Incidents / Total Hours Worked) * 200,000
# Handle potential division by zero if Total_Jam_Kerja is 0 or NaN
df_trir_calculated['TRIR'] = df_trir_calculated.apply(
    lambda row: (row['Total_Insiden'] / row['Total_Jam_Kerja']) * 200000 if row['Total_Jam_Kerja'] > 0 else 0,
    axis=1
)

print("TRIR data prepared and calculated successfully! First 5 rows:")
print(df_trir_calculated.head())
```

### Membuat Line Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_trir_calculated['Tahun'].astype(str) + '-' + df_trir_calculated['Bulan'].astype(str), df_trir_calculated['TRIR'], marker='o', linestyle='-', color='b')
plt.title('Monthly Total Recordable Incident Rate (TRIR) 2022-2024', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('TRIR', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Tambahkan trendline dan marker event

```python
plt.figure(figsize=(14, 7))
months_labels = df_trir_calculated['Tahun'].astype(str) + '-' + df_trir_calculated['Bulan'].astype(str)

plt.plot(months_labels, df_trir_calculated['TRIR'], marker='o', linestyle='-', color='b', label='Monthly TRIR')

# Add a trendline using a rolling mean (e.g., 3-month rolling average)
df_trir_calculated['TRIR_Rolling_Mean'] = df_trir_calculated['TRIR'].rolling(window=3, center=True).mean()
plt.plot(months_labels, df_trir_calculated['TRIR_Rolling_Mean'], linestyle='--', color='purple', label='3-Month Avg. Trendline')

# Add example event markers (e.g., for specific high TRIR months or policy changes)
# Replace these with actual event dates if you have them
event_months_indices = [7, 19, 31] # Example indices for events (adjust as needed)
event_names = ['Safety Campaign Launch', 'New Policy Implemented', 'Major Training Session']

for i, idx in enumerate(event_months_indices):
    if idx < len(months_labels):
        plt.axvline(x=months_labels.iloc[idx], color='gray', linestyle=':', lw=1.5, label=f'Event: {event_names[i]}' if i == 0 else "") # Label only once for legend
        plt.text(months_labels.iloc[idx], plt.ylim()[1]*0.95, f' {event_names[i]}', rotation=90, va='top', ha='right', fontsize=9, color='gray')

plt.title('Monthly Total Recordable Incident Rate (TRIR) 2022-2024 with Trendline and Events', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('TRIR', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

### Severity Rate (SR)

```python
# Persiapkan Data SR (Severity Rate)

# 1. Hitung Total_Hari_Kerja_Hilang dari df_insiden
df_lost_workdays_monthly = df_insiden.groupby(['Tahun', 'Bulan'])['Hari_Kerja_Hilang'].sum().reset_index()
df_lost_workdays_monthly = df_lost_workdays_monthly.rename(columns={'Hari_Kerja_Hilang': 'Total_Hari_Kerja_Hilang'})

# 2. Gabungkan dengan Total_Jam_Kerja dari df_kpi
df_sr_calculated = pd.merge(
    df_lost_workdays_monthly,
    df_kpi[['Tahun', 'Bulan', 'Total_Jam_Kerja']],
    on=['Tahun', 'Bulan'],
    how='left'
)

# 3. Hitung SR: (Total_Hari_Kerja_Hilang / Total_Jam_Kerja) * 200,000
df_sr_calculated['SR'] = df_sr_calculated.apply(
    lambda row: (row['Total_Hari_Kerja_Hilang'] / row['Total_Jam_Kerja']) * 200000 if row['Total_Jam_Kerja'] > 0 else 0,
    axis=1
)

print("Data SR prepared and calculated successfully! First 5 rows:")
print(df_sr_calculated.head())
```

### Membuat Line Chart SR

```python
plt.figure(figsize=(12, 6))
plt.plot(df_sr_calculated['Tahun'].astype(str) + '-' + df_sr_calculated['Bulan'].astype(str), df_sr_calculated['SR'], marker='o', linestyle='-', color='r')
plt.title('Monthly Severity Rate (SR) 2022-2024', fontsize=16)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Severity Rate (SR)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Gabungkan TRIR dan SR

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ensure fonts are set if not already in the session
try:
    plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei', 'DejaVu Sans', 'sans-serif']
    plt.rcParams['axes.unicode_minus'] = False
except KeyError:
    print("Warning: Could not set preferred fonts. Plotting with default Matplotlib fonts.")

fig, ax1 = plt.subplots(figsize=(16, 8))

# Prepare months labels with 'Month-YY' format
months_labels = df_trir_calculated['Bulan'].astype(str) + '-' + df_trir_calculated['Tahun'].astype(str).str[-2:]

# Calculate rolling means if not already present (ensure dataframes are updated)
if 'TRIR_Rolling_Mean' not in df_trir_calculated.columns:
    df_trir_calculated['TRIR_Rolling_Mean'] = df_trir_calculated['TRIR'].rolling(window=3, center=True).mean()
if 'SR_Rolling_Mean' not in df_sr_calculated.columns:
    df_sr_calculated['SR_Rolling_Mean'] = df_sr_calculated['SR'].rolling(window=3, center=True).mean()

# Plot SR on the primary Y-axis (ax1)
ax1.plot(months_labels, df_sr_calculated['SR'], marker='x', linestyle='-', color='r', label='Monthly SR')
ax1.plot(months_labels, df_sr_calculated['SR_Rolling_Mean'], linestyle='--', color='darkred', label='SR 3-Month Avg.')
ax1.set_xlabel('Month-Year', fontsize=12)
ax1.set_ylabel('Severity Rate (SR)', color='r', fontsize=12)
ax1.tick_params(axis='y', labelcolor='r')

# Create a secondary Y-axis for TRIR (ax2)
ax2 = ax1.twinx()
# Plot TRIR on the secondary Y-axis (ax2)
ax2.plot(months_labels, df_trir_calculated['TRIR'], marker='o', linestyle='-', color='b', label='Monthly TRIR')
ax2.plot(months_labels, df_trir_calculated['TRIR_Rolling_Mean'], linestyle='--', color='darkblue', label='TRIR 3-Month Avg.')
ax2.set_ylabel('Total Recordable Incident Rate (TRIR)', color='b', fontsize=12)
ax2.tick_params(axis='y', labelcolor='b')

# Adjust the limits of ax2 to match the ratio of ax1
# Find the maximum values to set the ratio
max_sr_value = df_sr_calculated['SR'].max()
max_trir_value = df_trir_calculated['TRIR'].max()

if max_sr_value > 0 and max_trir_value > 0:
    ratio = max_sr_value / max_trir_value
    # Set ax2 (TRIR) limits to match the ratio of ax1 (SR) limits
    # Ensure both start from 0
    ax1_ylim_max = ax1.get_ylim()[1] # Get current max of ax1
    ax2_ylim_max = ax1_ylim_max / ratio # Scale ax2 max based on ratio
    ax2.set_ylim(0, ax2_ylim_max)


# Add annotation for the last TRIR value
ax2.text(months_labels.iloc[-1], df_trir_calculated['TRIR'].iloc[-1],
         f'TRIR: {df_trir_calculated['TRIR'].iloc[-1]:.2f}',
         color='blue', ha='left', va='center', fontweight='bold')

# Add annotation for the last SR value
ax1.text(months_labels.iloc[-1], df_sr_calculated['SR'].iloc[-1],
         f'SR: {df_sr_calculated['SR'].iloc[-1]:.2f}',
         color='red', ha='left', va='center', fontweight='bold')

plt.title('Monthly TRIR and SR Trend (2022-2024)', fontsize=16, fontweight='bold')
plt.xticks(rotation=90, fontsize=10)
ax1.tick_params(axis='y', labelsize=10)
ax2.tick_params(axis='y', labelsize=10)
ax1.grid(True, linestyle='--', alpha=0.7)

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
import pandas as pd

# Ensure nltk stopwords are downloaded
try:
    stopwords.words('indonesian')
except LookupError:
    nltk.download('stopwords')

# Combine all incident descriptions into a single string
all_descriptions = ' '.join(df_insiden['Deskripsi_Insiden'].dropna().astype(str))

# Define Indonesian stopwords
indonesian_stopwords = set(stopwords.words('indonesian'))

# Add some common words that might not be useful for analysis
custom_stopwords = {'di', 'yang', 'dan', 'terjadi', 'pada', 'saat', 'dengan', 'karena', 'akibat', 'area', 'oleh', 'namun', 'untuk'}
all_stopwords = indonesian_stopwords.union(custom_stopwords)

# Generate the word cloud
wordcloud = WordCloud(width=1000, height=500, background_color='white',
                      stopwords=all_stopwords, min_font_size=10).generate(all_descriptions)

# Display the word cloud
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud: Deskripsi Insiden', fontsize=20)
plt.show()

# Get word frequencies
from collections import Counter
import re

# Clean text for frequency counting: lowercase, remove punctuation, split into words
words = re.findall(r'\b\w+\b', all_descriptions.lower())

# Filter out stopwords and single-character words
filtered_words = [word for word in words if word not in all_stopwords and len(word) > 1]

# Count word frequencies
word_counts = Counter(filtered_words)

print("\nTop 20 kata yang paling sering muncul:")
for word, count in word_counts.most_common(20):
    print(f"- {word}: {count}")
```
